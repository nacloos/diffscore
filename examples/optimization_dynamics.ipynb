```python
X = neural_data.values
print(X.shape)
# conditions = [
#     {"category": cat, "object": obj} for cat, obj in zip(neural_data["category_name"].values, neural_data["object_name"].values)
# ]
# convert to numerical
categories = np.unique(neural_data["category_name"].values)
objects = np.unique(neural_data["object_name"].values)
conditions = [
    {
        "category": np.where(categories == cat)[0][0],
        "object": np.where(objects == obj)[0][0]
    } 
    for cat, obj in zip(neural_data["category_name"].values, neural_data["object_name"].values)
]
print(conditions)

from diffscore.analysis import decoder_logistic

decoder_logistic(X, [cond["category"] for cond in conditions], penalty="l2")
```
